```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.barh(data['Estimated Unemployment Rate (%)'],data['Estimated Labour Participation Rate (%)'],color='green')
plt.xlabel('estimated unemployment rate')
plt.ylabel('estimated labour participation rate')
plt.title('estimated unemployment rate')
plt.show()
```

```python
plt.hist(data['Estimated Unemployment Rate (%)'],bins=20,color='red')
plt.xlabel('estimated unemployment')
plt.ylabel('frequency')
plt.title('Distribution of Estimated Unemployment rate')
plt.show()
```

```python
plt.figure(figsize=(8,6))
data['Estimated Unemployment Rate (%)'].plot(kind='density',color='blue')
plt.title('Density Plot of Unemployment')
plt.xlabel('Estimated Unemployment')
plt.ylabel('Density')
plt.show()
```
